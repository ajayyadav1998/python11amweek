```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from decimal import Decimal
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tip = sns.load_dataset('tips')
```

```python
tip.head()
```

```python
tip.info()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='sex',data=tip,palette='Greens')
plt.show()
```

```python
color = tip.sex.map({'Male':'red','Female':'green'})
```

```python
color
```

```python
plt.figure(figsize=(10,5))
plt.scatter(tip.total_bill,tip.tip,color=color)
plt.text(5,6,s='Male',color='red',fontsize=15)
plt.text(6,7,s='Female',color='green',fontsize=15)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
plt.figure(figsize=(15,7),facecolor='cyan')
sns.scatterplot(x='total_bill',y='tip',data=tip,hue='sex',size='smoker',palette='Set1')
plt.title('total_bill vs tip')
plt.show()
```
