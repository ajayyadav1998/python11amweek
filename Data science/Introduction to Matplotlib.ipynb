```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)   # number of row , 
plt.plot(x,y,'y--')
plt.title('first graph')

plt.subplot(1,3,2)
plt.plot(y,x,'g-.')
plt.title('second graph')

plt.subplot(1,3,3)
plt.plot([1,2,3,4,5],[1,4,9,16,25],'m-.')
plt.title('3rd graph')

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10,5),facecolor='seagreen')

axes1 = fig.add_axes([0,0,0.8,0.8]) #left,bottom,width,height
axes2 = fig.add_axes([0.2,0.4,0.23,0.25])

axes1.plot(x,y,'y--')
axes1.set_title('large graph')
axes1.set_xlabel('x-axis',fontdict={'color':'k','fontsize':15})
axes1.set_ylabel('y-axis',fontdict={'color':'k','fontsize':15})

axes2.plot(y,x,'k-.')
axes2.set_title('smaller graph')
axes2.set_xlabel('x-axis',fontdict={'color':'r','fontsize':15})
axes2.set_ylabel('y-axis',fontdict={'color':'r','fontsize':15})

plt.show()
```

```python
angles = [np.deg2rad(var) for var in range(361)]
```

```python
print(plt.style.available)
```

```python
plt.figure(figsize=(10,5),facecolor='seagreen')
plt.style.use('dark_background')
plt.plot(angles,np.sin(angles),'r',label='$Sin \Theta$',linestyle='-.')
plt.plot(angles,np.cos(angles),'g',label='$Cos \Theta$',linestyle='-.')
plt.grid()
plt.title('plot of sin and cos')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='seagreen')
plt.style.use('seaborn-darkgrid')
plt.plot(angles,np.sin(angles),'r',label='$Sin \Theta$',linestyle='-.')
plt.plot(angles,np.cos(angles),'g',label='$Cos \Theta$',linestyle='-.')
plt.title('plot of sin and cos')
plt.legend()
plt.show()
```

```python
x = np.linspace(-1,1,100)
```

```python
plt.figure(figsize=(10,5),facecolor='#800080')
plt.rcParams['text.color'] = 'pink'
plt.style.use('dark_background')
plt.plot(x,x,'y',label='X',linewidth=2)
plt.plot(x,x**2,'m',label='$X^2$',lw=2)
plt.plot(x,x**3,'g',label='$X^3$',lw=2)
plt.plot(x,x**4,'pink',label='$X^4$',lw=2)
plt.title('Awsome plot')
plt.xlabel('X-axis',fontdict={'color':'pink','fontsize':12})
plt.ylabel('Y-axis',fontdict={'color':'pink','fontsize':12})
plt.grid()
plt.legend()
plt.show()
```

```python
plt.style.use('default')
players=['bhargav','bhavik','yash','sanket','hiren','vismay']
runs=[453,456,345,345,665,675]
color=['red','pink','green','yellow','magenta','violet']
plt.figure(figsize=(10,5))
plt.bar(x=players,height=runs,color=color)
plt.title('runs of the players')
plt.xlabel('x-axis',fontsize=15)
plt.ylabel('y-axis',fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
country=['india','us','china','canada','africa','uk']
output=[21,32,34,56,45,65]  #in crores
explode=[0,0,0,0,0,0.2]
color=['red','pink','k','y','m','g']
plt.figure(figsize=(10,5),facecolor='seagreen')
plt.pie(x=output,labels=country,autopct='%.1f%%',explode=explode,shadow=True,colors=color,textprops={'color':'w','fontsize':12})
plt.show()
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2)

ax[0,0].plot(x,y,'r--')
ax[0,0].set_title('1st plot')
ax[0,0].set_xlabel('x-axis')
ax[0,0].set_ylabel('y-axis')

ax[0,1].plot(x,y,'y-.')
ax[0,1].set_title('2st plot')
ax[0,1].set_xlabel('x-axis')
ax[0,1].set_ylabel('y-axis')
                   
ax[1,0].plot([1,2,3,4,5],[1,4,9,16,25],'g-+')
ax[1,0].set_title('3st plot')
ax[1,0].set_xlabel('x-axis')
ax[1,0].set_ylabel('y-axis')
                   
ax[1,1].plot([1,2,3,4,5],[6,5,7,4,8],'k-.')
ax[1,1].set_title('4st plot')
ax[1,1].set_xlabel('x-axis')
ax[1,1].set_ylabel('y-axis')                   


plt.suptitle('various subplots')

plt.tight_layout()
```
