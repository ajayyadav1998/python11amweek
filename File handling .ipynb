```python
file = open('tbale.txt')
data = file.read()
file.close()
```

```python
data
```

```python
print(data)
```

```python
data = data.split('\n')
```

```python
data
```

```python
table = list(map(lambda string : string.split(),data))
table
```

```python
q = []
r = []
s = []

for a,b,c,d in table:
    print(a,b,c,d,sep='\t')
    q.append(b)
    r.append(c)
    s.append(d)
```

```python
q.remove('math')
r.remove('phy')
s.remove('chem')
```

```python
t = q + r + s
print(t)
```

```python
ms = reduce(lambda x,y : int(x)+int(y),t)
ms
```
